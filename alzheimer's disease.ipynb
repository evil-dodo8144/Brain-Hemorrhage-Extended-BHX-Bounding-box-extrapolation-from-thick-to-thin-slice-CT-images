```python
import cv2
import numpy as np
import pandas as pd
img = cv2.imread("example_annotated_images.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.resize(gray, (224, 224))
gray = gray / 255.0
features = gray.flatten()
df = pd.DataFrame(features, columns=["pixel_value"])
df.to_csv("brain_ct_pixels.csv", index=False)
print("Saved brain_ct_pixels.csv")
```
